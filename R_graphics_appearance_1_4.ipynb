В ggplot2 есть разные способы настройки цвета. Если вы расположите аргумент col (на самом деле он color или colour, но все три варианта ggplot одинаково поймёт) вне функции aes, эта конструкция будет работать примерно так же, как в базовой графике, за одним важным исключением.

# Настройка цвета продолжение

 Если вы расположите аргумент colo(u)r или fill внутри функции aes(), той самой, которую мы кормим информацией про x и y, то ей нужно дать на вход фактор, и тогда она определит цвета для каждого уровня фактора и автоматически добавит легенду.

```r
ggplot(iris, aes(Species, Petal.Width, fill = Species)) + geom_boxplot()
```

Параметры графика можно настроить внутри самой функции plot(), которая строит основной график, или вынести в отдельную функцию par(), которая должна быть расположена перед plot(). Вот как, например, изменить тип, размер и цвет символов в графике:

plot(cars, pch = 19, cex=0.5, col = "red4")

Полный список изменяемых параметров можно посмотреть с помощью справки к функции par(): ?par. Также краткое изложение графических параметров есть в разделе Details справки к функции plot.default(): ?plot.default.
