# Boxplot
Взаимосвязь между факторной переменной (категориальной переменной, группами)  и переменной числовой, то обычно самый удачный выбор — это «коробка с усами» (boxplot)

```r
boxplot(iris$Petal.Length ~ iris$Species)
```

```r
ggplot(iris, aes(Species, Petal.Length)) + geom_boxplot()
```

```r
ggplot(iris, aes(y=Petal.Length, x=Species)) + geom_boxplot()
```

# Barplot

 взаимосвязи между факторной и числовой переменной может пригодиться столбчатая диаграмма (barplot).

```r
# игрушечные данные
sampling.result <- data.frame(sex = c("male", "female", "juvenile"), quantity = c(156, 209, 1506))
```

```r
barplot(sampling.result$quantity ~ sampling.result$sex)
```

```r
ggplot(sampling.result, aes(sex, quantity)) + geom_col()
```

```r
ggplot(sampling.result, aes(sex, quantity)) + geom_bar(stat = "identity")
```

# Histogram

Для отображения распределения в одном числовом векторе полезна гистограмма (histogram). Суть этого подхода в том, чтобы распределить значения числового вектора в группы близких значений и отобразить на графике количество значений в каждой группе.

```r
hist(iris$Petal.Length)
```

```r
hist(iris$Petal.Length, breaks = 40)
```

```r
ggplot(iris, aes(Petal.Length)) + geom_histogram()
```

```r
ggplot(iris, aes(Petal.Length)) + geom_histogram(bins = 40)
```

```r
ggplot(iris, aes(Petal.Length)) + geom_histogram(binwidth = 1)
```

# Line plot

Если речь идёт о динамике изменения переменной, бывает полезна линейная диаграмма (line plot).

```r
subject1 <- Indometh[Indometh$Subject == 1,]
```

```r
plot(subject1$time, subject1$conc, type = "l")
```

```r
subject1 <- Indometh[Indometh$Subject == 1,]
subject2 <- Indometh[Indometh$Subject == 2,]
plot(subject1$time,subject1$conc, type = "l")
lines(subject2$time,subject2$conc)
```

```r
ggplot(subject1, aes(time, conc)) + geom_line()
```
